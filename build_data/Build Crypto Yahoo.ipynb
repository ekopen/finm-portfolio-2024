```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
TICKS = [
    'BTC-USD',
    'ETH-USD',
    'SOL-USD',
    'DOGE-USD',
]

FLDS = ['quoteType','shortName','volume']
```

```python
info = pd.DataFrame(index=TICKS,columns=FLDS)
info.index.name = 'ticker'
for tick in info.index:
    temp = yf.Ticker(tick).get_info()

    for fld in FLDS:
        if fld in temp.keys():
            info.loc[tick,fld] = temp[fld]
```

```python
prices = yf.download(TICKS, start="2017-01-01", end="2024-10-31")['Adj Close'][info.index]
prices.index = prices.index.tz_localize(None)
prices.dropna(inplace=True)

rets = prices.pct_change().dropna()
```

```python
with pd.ExcelWriter('../data/crypto_data.xlsx') as writer:  
    info.to_excel(writer, sheet_name= 'descriptions')
    rets.to_excel(writer, sheet_name='total returns')
    prices.to_excel(writer, sheet_name='prices')
```
