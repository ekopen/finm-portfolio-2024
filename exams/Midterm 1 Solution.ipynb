# <span style="color:red">Solution 1</span>

### <span style="color:red">1.1.</span>

**TIPS is found to have correlation of 0% to `IEF` and 0% to `SPY`:**

If restricted to a long-only portfolios, correlation of 0% would for sure increase TIPS diversification characteristic and therefore increase its weight in the GMV and tangency portfolio. In a long-only portfolio, lower correlation with other assets, higher expected return, and lower volatility (the last two are directly related to Sharpe) are the main estimates responsible for defining (and more specifically increasing) weights in optimal MV portfolios. In summary, by lowering TIPS correlation with SPY and IEF we can ensure bigger absolute value of Sharpe in the tangency portfolio.

In the case study as well as in our homework (even though we are not restricted to long-only portfolio), TIPS, IEF and SPY present positive weights, which leads us to conclude that a decrease in TIP correlation with the specified other assets should improve the MV allocation and increase the set of risk-return possible allocations.

Extra:

In some cases, if not restricted to long-only portfolios, the expected output is not straight-forward. For instance, if TIPS currently had negative weights in the tangency portfolio, decreasing its correlation with other assets to 0% (all else constant) would decrease its diversification benefit (shorting an asset with highly positive correlation is equal to buying an asset with highly negative correlation). The result could also be detrimental even if TIPS has positive weight but IEF and SPY have negative weights in the tangency portfolio, since the change would reduce the diversification benefit of IEF and SPY.

**TIPS is found to have correlation of 100% to `IEF`:**

If TIPS had a correlation of 100% with IEF, the covariance matrix would not be invertible, not allowing for a analytical solution when deriving the tangency and GMV portfolios. The theoretical solution, in the limit, would tell you to infinitely short the asset with the lowest Sharpe between the two to infinitely buy the asset with the highest Sharpe. In the tangency portfolio constrained optimization, the solution would put IEF weight in its lower bound (considering that it has lower Sharpe than TIPS) and TIPS on its upper bound.

Alternatively, you could drop one of the assets. In this scenario, it would make sense to drop IEF, again due to TIPS higher Sharpe.

**Would it be possible for TIPS to have been found to have 0% correlation to every other asset in `homework 1`? Explain.**

Yes, it is mathematically possible even though economically improbable. Such a covariance matrix would show that TIP is not affected by other market factors/events that generate consequences to all other market broad indexes.

In this improbable case, TIPS would have the highest diversification benefit and we should expect the tangency portfolio and the GMV portfolio to allocate relevant weight to it.

Curiosity:

In this scenario, running the optimization with all the other assets but TIPS and creating a portfolio which would later be optimized with TIPS would yield the same result as optimizing the solution with all the assets (including TIPS) directly.

# <span style="color:red">Solution 4</span>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fx_carry_rets = pd.read_excel('midterm_1_data.xlsx', sheet_name='fx carry returns', index_col=0)
jpy_rets = fx_carry_rets['JPY']
```

### <span style="color:red">4.1.</span>

```python
# JPY VaR empirical (expanding methodology)
jpy_var_emp_exp = jpy_rets.expanding().quantile(0.01)

jpy_var_emp_exp.plot(figsize=(15, 7), title='JPY FX Carry Empirical VaR (1%)', xlabel='Date', ylabel='VaR (1%)'); plt.show()
```

```python
# VaR as of final date
jpy_var_emp_exp.iloc[-1]
```

### <span style="color:red">4.2.</span>

```python
# JPY sigma (population mean of 0, rolling methodology with a 233-day window)
sigma_roll = np.sqrt((jpy_rets ** 2).rolling(window=233).mean()) 
# Or jpy_rets.rolling(window=window_size,center=False).apply(lambda x: np.sqrt(np.sum(x**2)/ window_size))
# JPY VaR (normal distribution methodology)
jpy_var_norm_roll = -2.33 * sigma_roll

jpy_var_norm_roll.plot(figsize=(15, 7), title='JPY FX Carry Rolling Normal VaR (1%)', xlabel='Date', ylabel='VaR (1%)'); plt.show()
```

```python
# VaR as of final dates
jpy_var_norm_roll.iloc[-3:]
```

```python
(jpy_rets.rolling(window=233,center=False).apply(lambda x: np.sqrt(np.sum(x**2)/ 233)).shift(1)*-2.33).iloc[-3:]
```

### <span style="color:red">4.3.</span>

```python
lbda = 0.94
# JPY sigma (population mean of 0, ewma methodology)
sigma_ewma = np.sqrt((jpy_rets ** 2).ewm(alpha=1 - lbda, adjust=False).mean())
# JPY VaR (normal distribution methodology)
jpy_var_norm_ewma = -2.33 * sigma_ewma

jpy_var_norm_ewma.plot(figsize=(15, 7), title='JPY FX Carry EWMA Normal VaR (1%)', xlabel='Date', ylabel='VaR (1%)'); plt.show()
```

```python
# VaR as of final dates
jpy_var_norm_ewma.iloc[-3:]
```

```python
jpy_var_norm_ewma.shift(1).iloc[-3:]
```

__Note:__ We could have multiple EWMA methodologies that should be fine. The results shown above reflect the formulas shown in the discussion notebook titled "Value-at-Risk" (https://github.com/MarkHendricks/finm-portfolio-2024/blob/main/discussions/D.3.%20Value-at-Risk.ipynb).

See https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ewm.html for a more detailed discussion of parameters. In particular, setting `adjust=False` is equivalent to the methodology in the discussion notebook.

### <span style="color:red">4.4.</span>

```python
jpy_var_estimates = pd.concat([jpy_var_norm_ewma, jpy_var_norm_roll, jpy_var_emp_exp], axis=1, keys=['Normal (EWMA)', 'Normal (Rolling)', 'Empirical (Expanding)'])
jpy_var_estimates.plot(figsize=(15, 7), legend=True, title='VaR (1%) Estimation Comparisons', xlabel='Date', ylabel='VaR (1%)')
```

Evidently, the EWMA approach changes the most dynamically, which could help us quickly change our VaR estimates as market conditions change. On the other hand, putting too much weight in recent values can be a myopic strategy as market conditions can change drastically in very little time.

The rolling approach has similar characteristics to the EWMA approach, and the window length could be fined tuned to more closely match the EWMA methodology. In any case, the rolling approach is susceptible to large jumps as values enter/exit the window.

The expanding empirical approach has the benefit of including all data points, which is likely necessary due to the poor statistical power of estimating the 1% quantile of a distribution in this manner. If we trust our estimates, or our return distribution is highly non-normal, then the empirical method can better capture the dynamics of the tail of our distribution. However, using all data points means that the estimates are less responsive to market conditions and estimates from this approach can be sensitive to sample size.

### <span style="color:red">4.5.</span>

We like to use the hit-rate to judge the performance a VaR model, which tells us what proportion of the time our returns breach our VaR threshold. A good VaR (1%) estimator should achieve a hit rate close to 1%.

In the previous questions, little attention was paid to the indexing of our VaR estimates. In fact, an index at time $t$ reflected our estimate for VaR based on data up to *and including* time $t$. We must now pay close attention to ensure we have no lookahead bias in our VaR estimates, and calculate our hit rate accordingly.

To make this comparison fair, we might only want to estimate the hit rate for times at which we have a non-NaN estimate for each methodology. We will do both here for completeness.

```python
# The aligned data, such that each VaR estimate applies to the JPY returns at the same time index (no lookahead bias)
var_estimates_t = jpy_var_estimates.shift(1)
pd.concat([jpy_rets.rename('Returns'), var_estimates_t], axis=1)
```

```python
# Hit rates, excluding times where we have NaN estimates for any estimator
var_estimates_t.dropna().apply(lambda x: x >= jpy_rets.loc[x.index]).mean()
```

```python
# Hit rates, including all available data for that estimator
var_estimates_t.apply(lambda x: x.dropna() > jpy_rets.loc[x.dropna().index]).mean()
```

We find that all of the hit rates are significantly above 1%, meaning we are underestimating our risk with these estimates. Out of the three, the EWMA and expanding methods perform best. Using the same data for each, the empirical approach performs the greatest.

__Note__: Hit ratio of 0.0199 is also accepted for EWMA as different methodologies can be used to derive EWMA VaR and can generate small differences in answers. 0.0207 is accepted for rolling estimates hit rate due to different NaN estimates. Common mistakes involve forgetting to shift index (Wrong answer for rolling - 0.199, Expanding - 0.23, EWMA - 0.115).
